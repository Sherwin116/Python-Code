```python
# Importing the packages

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None
```

```python
# Reading the data 

df = pd.read_csv(r'C:\Users\Admin\Desktop\movies.csv')
```

```python
# Looking at the data 

df
```

```python
# Checking is there's any missing data 

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Dropping the missing values

df = df.dropna()
```

```python
# Data types for the columns

df.dtypes
```

```python
# Changing data type of the columns to make it look organised 

df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')

df['votes'] = df['votes'].astype('int64')

df['runtime'] = df['runtime'].astype('int64')
```

```python
df
```

```python
df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# Dropping Duplicates

df.drop_duplicates()
```

```python
# Scatter plot with budget vs gross 

plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs Gross Earnings')

plt.xlabel('Gross Earnings')

plt.ylabel('Budget for film')

plt.show()
```

```python
df.head()
```

```python
# plot budget vs gross using seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color":"purple"}, line_kws={"color":"black"})
```

```python
# Looking at correlation 

df.corr(method='pearson')
```

```python
correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Metric for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized
        
```

```python
correlation_matrix = df_numerized

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Metric for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```
